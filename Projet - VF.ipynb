```python
df_nutri_1.head(3)
```

```python
X = df_nutri_1[features_col]
X
```

```python
X.columns
```

```python
y = df_nutri_1[target_col]
y.head()
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# ------------------------------------------------------------------------------------

# 1- Logistic Regression :

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score

lr = LogisticRegression()
lr.fit(X_train, y_train)

acc_lr = lr.score(X_test, y_test)
print (f"Accuracy = {round(acc_lr * 100, 2)}%")

print(f"F1-score Logistic Regression on train  {round(f1_score(y_train, lr.predict(X_train), average = 'micro') * 100, 2)}%")
print(f"F1-score Logistic Regression on test = {round(f1_score(y_test,  lr.predict(X_test), average = 'micro') * 100, 2)}%")
```

# ------------------------------------------------------------------------------------

# 2- Tree :

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier() 

dt.fit(X_train, y_train)

y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc_dt = accuracy_score(y_test, y_pred)
print (f"Accuracy Tree = {round(acc_dt * 100, 2)}%")

print(f"F1-score Tree on train  {round(f1_score(y_train, dt.predict(X_train), average = 'micro') * 100, 2)}%")
print(f"F1-score Tree on test = {round(f1_score(y_test,  dt.predict(X_test), average = 'micro') * 100, 2)}%")
```

```python
## Importance de chaque feature :
```

```python
features_imp = dt.feature_importances_
features_imp
```

```python
acc_rf = accuracy_score(y_test, y_pred)
print (f"Accuracy Random Forest = {round(acc_rf * 100, 2)}%")

print(f"F1-score Random Forest on train  {round(f1_score(y_train, rf.predict(X_train), average = 'micro') * 100, 2)}%")
print(f"F1-score Random Forest on test = {round(f1_score(y_test,  rf.predict(X_test), average = 'micro') * 100, 2)}%")
```
